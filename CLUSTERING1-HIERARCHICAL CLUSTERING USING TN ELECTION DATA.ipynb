## Objective : Learning the Implementation of Ward's method

## Question :  Implement Ward's method for TN election data and identify the number of clusters and also the members in the cluster -1 

# Step 1 : Data Preperation

```python
import pandas as pd
df=pd.read_csv("tnelections.csv")
df
```

```python
X=df.iloc[:,1:6]
X
```

# Step 2 : Implementation

```python
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(6,6))
plt.title("Ward's dendrogram for TN election Data")
Dendrogram=sch.dendrogram(sch.linkage(X,method='ward'))
```

## Conclusions :
## There are three clusters of sizes 3, 2 and 6
## The first cluster has objects 3,0,2
## Members of the first cluster are Thiruvallur, Gummidipoondi, Tiruttani
