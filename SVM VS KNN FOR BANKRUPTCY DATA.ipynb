## Objective : Learning the process of comparing two classifiers

## Problem : Compare SVC with knn-classification with settings of your choice using bankruptcy data

## Step 1 : Data Preperation

```python
import pandas as pd
df=pd.read_csv("bankruptcy.csv")
df
```

## Step 2 : Slicing of data and target

```python
import numpy as np
X=df[["X1","X2","X3","X4"]] # data part
y=df[["cl"]] # target part
# X and y are data frames
```

# Step 3 : Importing SVC module and fitting for the given data

```python
from sklearn.svm import SVC
clf = SVC(gamma='auto')
y=np.array(y) # convert y into an array
y_rs=y.reshape((46,)) # reshaped into one dimensional array
clf.fit(X, y)
```

```python
y_predicted=clf.predict(X)
y=np.array(y)
y_rs=y.reshape((-1,))
```

```python
y
```

```python
y_predicted
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_rs,y_predicted)
```

#  Step 4 : Getting the validity metrics for SVC

```python
from sklearn import metrics
accuracy=metrics.accuracy_score(y_rs,y_predicted)
precision=metrics.precision_score(y_rs,y_predicted)
recall=metrics.recall_score(y_rs,y_predicted)
fscore=metrics.f1_score(y_rs,y_predicted)
print("\n The accuracy of support vector classifier is",np.round(100*accuracy,2),"%")
print("\n The precision of support vector classifier is",np.round(100*precision,2),"%")
print("\n The Recall of support vector classifier is",np.round(100*recall,2),"%")
print("\n The F-Score of support vector classifier is",np.round(100*fscore,2),"%")
```

# Step 5: knn classification

```python
from sklearn.neighbors import KNeighborsClassifier
k=8
clfk=KNeighborsClassifier(n_neighbors=k)
clfk.fit(X,y)
y_predk=clfk.predict(X)
```

```python
cnfknn=confusion_matrix(y_rs,y_predk)
cnfknn
```

```python
accuracy=metrics.accuracy_score(y_rs,y_predk)
precision=metrics.precision_score(y_rs,y_predk)
recall=metrics.recall_score(y_rs,y_predk)
fscore=metrics.f1_score(y_rs,y_predk)
print("\n The accuracy of k-nn classifier is",np.round(100*accuracy,2),"%")
print("\n The precision of k-nn classifier is",np.round(100*precision,2),"%")
print("\n The Recall of k-nn  is",np.round(100*recall,2),"%")
print("\n The F-Score of k-nn classifier is",np.round(100*fscore,2),"%")
```

# Conclusions : Knn has more accuracy , 
# better precision, recall and F-score
